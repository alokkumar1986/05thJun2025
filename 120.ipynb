```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

print(x)

plt.hist(x)

plt.title('Height frequency of 250 people')

plt.xlabel('Heights')

plt.ylabel('No. of People')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([10, 30, 5])
labels = np.array(['Male', 'Female', 'Others'])

colors = np.array(['green', 'red', 'b'])

explode= np.array([0.2, 0, 0])

plt.pie(x, labels=labels, colors=colors, explode=explode, shadow=True, radius=1.2 , startangle=0, autopct="%.1f%%")

plt.legend(title='Genders in our Class')

plt.title('The Gender distribution chart of our class')

plt.show()
```
